### Calcular distancias

```python
#Lo primero en clustering es calcular las distincias
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
data=pd.read_csv('C:/Users/Vic/Documents/CursoMachineLearning/Dataset/Movies/movies.csv', sep=';')
```

```python
#En el dataframe movies, no es necesario normalizar, puesto que el tema de las valoraciones ya estan normalizadas.
#todos estan en el rango 0 a 10
data
```

```python
#Excluyendo la columna 0 'user_id' no interesa para el calculo de distnacias.
movies=data.columns.values.tolist()[1:]
```

```python
movies
```

```python
#Calculando las distancias entre dos puntos.
#Las peliculas con las mismsa peliculas se van a verificar.
dd1=distance_matrix(data[movies],data[movies],p=1) #1=distancias manhattan
dd2=distance_matrix(data[movies],data[movies],p=2)  #2=distancia euclidia
dd10=distance_matrix(data[movies],data[movies],p=10) #10=cuanto mas grande sea p, menor es la distancia.
```

```python
dd1
```

```python
#Convertir un matriz de distancia a dataframe
def dm_to_df(dd,col_name):
    import pandas as pd
    return pd.DataFrame(dd,index=col_name, columns=col_name)
```

```python
#Matriz de distancias manhattan
dm_to_df(dd1,data['user_id'])
```

```python
#Matriz de distancias euclidian
dm_to_df(dd2,data['user_id'])
```

```python
#Matriz de distancias con valor mas grande
dm_to_df(dd10,data['user_id'])
```


### Metodos de enlace (Enlace simple)

```python
#Los metodos de enlace ayudan a enlazar o unir cluster.
#Se va reecrear manualmente el proceso de clustering.
```

```python
df=dm_to_df(dd1,data['user_id'])
df
```

```python
#En z, se va ir colocando los grupos y a que distnacia se encuentran.
#Colocar que grupos hemos unido y la distancia en la que se encuentran
#El valor mas pequeno es 0.7, se encuentra entre la fila 1 y 10. 
z=[]
```

```python
#Agregar la fila y columna

#Con la informacion del valor mas pequeno crear la columna y fila 11
df[11]=df[1]+df[10] #agregar la columna 11
df.loc[11]=df.loc[1]+df.loc[10]# agregar la fila 11

#Formar un nuevo cluster de dos elementos a partir de 1 y 10.
z.append([1,10,0.7,2]) #entre la 1 y 10 se ha encontrado el valor mas pequeno 0.7. Se forma un nuevo cluster
```

```python
df
```

```python
#Recualcular las distancias

#Luego de adicionar la columna 11, recalcular las distancias a todos Y cada uno de los punto.
for i in df.columns.values.tolist():
    df.loc[11][i]=min(df.loc[1][i],df.loc[10][i]) #encontrar el valor minimo y guardarlo en la colummna 11
    df.loc[i][11]=min(df.loc[i][1],df.loc[i][10]) #encontrar el valor minimo y guardarlo en la fila 11
    
    #print("i------->",i)
    #print("valores 1:",df.loc[1][i],df.loc[10][i])
    #print("min 1:",df.loc[11][i])
    #print("valores 2:",df.loc[i][1],df.loc[i][10])
    #print("min 2:",df.loc[i][11])
df
```

```python
#Eliminando filas 1 y 10
#En la 11 esta toda la informacion de la 1 y 10.

#Una ves recalculado eliminar la columna 1 y 10. 
#Toda la informacion esta en la fila 11 y recalculado
#En la columna 11 esta agrupa los valores de 1 y 10
df=df.drop([1,10]) #Borrar las filas 1 y 10
df=df.drop([1,10],axis=1) #Borrar las columnas 1 y 10
df
```

```python
#El valor minimo se encuentre en x,y
x=2
y=7
n=12 #representa la nueva fila y columna

df[n]=df[x]+df[y] 
df.loc[n]=df.loc[x]+df.loc[y]
z.append([x,y,df.loc[x][y],2]) 

#Recalcular las distancias a todos
for i in df.columns.values.tolist():
    df.loc[n][i]=min(df.loc[x][i],df.loc[y][i]) 
    df.loc[i][n]=min(df.loc[i][x],df.loc[i][y])
    
#Una ves recalculado eliminar la columna 1 y 10. 
df=df.drop([x,y])
df=df.drop([x,y], axis=1)
df
```

### Clusteriong Jerarquico

```python
#El Cluster Jerarquico no tiene la necesidad de conocer a priori el numero de cluster. 
#El algoritmo lo hace todo por nosotros.
#Pensado para la aglomeracion y asociacion de datos.

import matplotlib.pyplot as plt
#linkage, asigna un metodo de generacion de cluster
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
#linkage, calcula las distancias y la aplicacion del enlace. Genera automaticamente los cluster.
#linkage automaticamente genera los cluster y asigna los elementos a los cluster.
#ward(completo), average, single, son los metodos de enlace. Es un metodo para generar cluster dentro de un conjunto de datos.

#Z = linkage(data[movies],'ward')
#Z

#Ej. 
    #Se agarro el registro 0,9. 
    #El valor mas pequeno es 0.41231056. 
    #Se ha unidos dos elementos para formar un cluster
```

```python
#ward, es un metodo para generar cluster dentre de un conjunto de datos
Z = linkage(data[movies],'ward')
Z

#Distancias 0.41, 0.61, 2.16, 3.48
```

```python

plt.figure(figsize=(25,10))#tamano da la figura
plt.title('Dendrograma Jerarquico para Clustering')
plt.xlabel('Id de los usuario de netflix')
plt.ylabel('Distancia')
dendrogram(Z,leaf_rotation=90.,leaf_font_size=10.0)
plt.show()
```

```python
#avarage (o metodo promedio), es un metodo para generar cluster dentre de un conjunto de datos
Z = linkage(data[movies],'average')
Z
```

```python
plt.figure(figsize=(25,10))
plt.title('Dendograma Jerarquico para Clustering')
plt.xlabel('Id de los usuario de netflix')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=90.,leaf_font_size=10.0)
plt.show()
```

```python
#simple, es un metodo para generar cluster dentro de un conjunto de datos
Z = linkage(data[movies],'single')
Z
plt.figure(figsize=(25,10))
plt.title('Dendograma Jerarquico para Clustering')
plt.xlabel('Id de los usuario de netflix')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=90.,leaf_font_size=10.0)
plt.show()
```

```python
#Metodo de enlace=Single; matriz de distancia = cityblock (Manhattan)
Z = linkage(data[movies],method='single', metric='cityblock') #metrica cityblock = manhatan; 
Z
plt.figure(figsize=(25,10))
plt.title('Dendograma Jerarquico para Clustering')
plt.xlabel('Id de los usuario de netflix')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=90.,leaf_font_size=10.0)
plt.show()
```

```python
#Metodo de enlace=Single; matriz de distancia = minkowski
Z = linkage(data[movies],method='single', metric='minkowski') #metrica cityblock = manhatan; 
Z
plt.figure(figsize= (25,10))
plt.title('Dendograma Jerarquico para Clustering')
plt.xlabel('Id de los usuario de netflix')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=90.,leaf_font_size=10.0)
plt.show()
```

```python
#Metodo de enlace=Single; matriz de distancia = euclidean
Z = linkage(data[movies],method='single', metric='euclidean') #metrica cityblock = manhatan; 
Z
plt.figure(figsize=(25,10))
plt.title('Dendograma Jerarquico para Clustering')
plt.xlabel('Id de los usuario de netflix')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=90.,leaf_font_size=10.0)
plt.show()
```
