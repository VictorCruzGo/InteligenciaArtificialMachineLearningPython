```python
#Victor Cruz Gomez

#Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Leer el dataset reviews sentimients. 
dataset=pd.read_csv("C:/Users/Vic/Documents/CursoMachineLearning/Dataset/reviews_sentiment/reviews_sentiment.csv",sep=';')
```

```python
#Mostrar datos del dataset
dataset
```

```python
#Mostrar la cantidad de filas y columnas del dataset
dataset.shape
```

```python
#
x=dataset.iloc[:,[2,6]].values#wordcount,sentimientoValue
y=dataset.iloc[:,5].values#Star Rating
```

```python
pd.DataFrame(x).head()
```

```python
pd.DataFrame(y).head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
#from sklearn.preprocessing import MinMaxScaler
```

```python
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)

#scaler = MinMaxScaler()
#x_train=scaler.fit_transform(x_train)
#x_test=scaler.transform(x_test)
```

```python
pd.DataFrame(x_train).head()
#pd.DataFrame(x_train).describe()
```

```python
pd.DataFrame(x_test).head()
```

```python
#pd.DataFrame(y_train).head()
y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clasificacion=KNeighborsClassifier(n_neighbors=7)
clasificacion.fit(x_train,y_train)
```

```python
y_pred=clasificacion.predict(x_test)
pd.DataFrame(y_pred).head()
```

```python
pd.DataFrame(y_test).head()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
clasificacion.score(x,y)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
from matplotlib.colors import ListedColormap
```

```python
#x_train
#y_train
```

```python
x_set, y_set = x_train, y_train

x1, x2 = np.meshgrid(np.arange(start = x_set[:,0].min() - 1, stop = x_set[:,0].max() + 1, step=0.01),
                    np.arange(start = x_set[:,1].min() - 1, stop = x_set[:,1].max() + 1, step=0.01))

plt.contourf(x1,x2, 
            clasificacion.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),
             alpha = 0.20,cmap = ListedColormap(('red','green','blue','yellow','orange')))

plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())

#Pintar los puntos de verde y rojo
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j,0], x_set[y_set == j,1],
                c = ListedColormap(('red','green','blue','yellow','orange'))(i), label=j)
        

plt.title("KNN (Conjunto de Entrenamiento)")
plt.xlabel("wordcount")
plt.ylabel("sentimientoValue")
plt.legend()
plt.show()
```

```python
x_set, y_set = x_test, y_test
x1, x2 = np.meshgrid(np.arange(start = x_set[:,0].min() - 1, stop = x_set[:,0].max() + 1, step=0.01),
                     np.arange(start = x_set[:,1].min() - 1, stop = x_set[:,1].max() + 1, step=0.01))
plt.contourf(x1,x2, clasificacion.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),
            alpha = 0.75,cmap = ListedColormap(('red','green','blue','yellow','orange')))
plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j,0], x_set[y_set == j,1],
                c = ListedColormap(('red','green','blue','yellow','orange'))(i), label=j)
plt.title("KNN (Conjunto de Testing)")
plt.xlabel("Edad")
plt.ylabel("Sueldo Estimado")
plt.legend()
plt.show()
```
