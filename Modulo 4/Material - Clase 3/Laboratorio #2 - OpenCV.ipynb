```python
#Invertir el orden de los colores de BGR a RGB
#img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#plt.imshow(img_rgb)

img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

## RESIZE IMAGE

```python
#Tamano de la imagen
img_rgb.shape
#(alto,ancho,rgb)
```

```python
img.shape
```

```python
#plt.imshow(img)
plt.imshow(img)
```

```python
#Redimensionar imagen

w_ratio = 0.5
h_ratio = 0.5
```

```python
#new_img = cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
#plt.imshow(new_img)
new_img=cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
plt.imshow(new_img)
```

## FLIPPING IMAGE (voltear imagen)

```python
%matplotlib inline
```

```python
#0, flipping la imagen alrededor del eje x.
#>1, flipping la imagen alredero del eje y.
#<0, flipping la imagen en ambos ejes. Primero x, luego Y.

#new_img = cv2.flip(new_img,0)
#plt.imshow(new_img)

new_img=cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
#new_img = cv2.flip(new_img,1)
#plt.imshow(new_img)

new_img=cv2.flip(new_img,1)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(new_img,-1)
plt.imshow(new_img)
```
