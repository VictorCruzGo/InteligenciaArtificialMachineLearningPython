```python
#Importar modulos.
import numpy as np
import matplotlib.pyplot as plt
import cv2 #OpenCV
%matplotlib inline
```

```python
#Crear una imagen con fondo blanco de tamano 1024x1024
black_board = np.zeros(shape=(1024,1024,3),dtype=np.int16)
```

```python
#Mostrar forma de la imagen
black_board.shape
```

```python
#Mostrar la imagen
plt.imshow(black_board)
```

```python
#Dibujar lineas bordes
#lineas horizontales superior
cv2.line(black_board,pt1=(20,20),pt2=(1004,20),color=(255,0,0),thickness=3)
cv2.line(black_board,pt1=(40,40),pt2=(984,40),color=(255,0,0),thickness=3)

#lineas horizontales inferior
cv2.line(black_board,pt1=(20,1004),pt2=(1004,1004),color=(255,0,0),thickness=3)
cv2.line(black_board,pt1=(40,984),pt2=(984,984),color=(255,0,0),thickness=3)

#lineas verticales izquierdo
cv2.line(black_board,pt1=(20,20),pt2=(20,1004),color=(255,0,0),thickness=3)
cv2.line(black_board,pt1=(40,40),pt2=(40,984),color=(255,0,0),thickness=3)

#lineas verticales derecho
cv2.line(black_board,pt1=(1004,20),pt2=(1004,1004),color=(255,0,0),thickness=3)
cv2.line(black_board,pt1=(984,40),pt2=(984,984),color=(255,0,0),thickness=3)

plt.imshow(black_board)
```

```python
#Dibujar diagonales
#linea diagonal principal
cv2.line(black_board,pt1=(0,0),pt2=(1024,1024),color=(255,0,0),thickness=3)

#lineas diagonal secundaria
cv2.line(black_board,pt1=(0,1024),pt2=(1024,0),color=(255,0,0),thickness=3)

plt.imshow(black_board)
```

```python
#Dibujar circulos centrales
#circulo 1
cv2.circle(black_board,center=(512,512), radius=50, color=(255,0,0), thickness=20)

#circulo 2
cv2.circle(black_board,center=(512,512), radius=100, color=(255,0,0), thickness=3)

#circulo 3
cv2.circle(black_board,center=(512,512), radius=150, color=(255,0,0), thickness=3)

#circulo 4
cv2.circle(black_board,center=(512,512), radius=200, color=(255,0,0), thickness=15)

plt.imshow(black_board)
```

```python
#Dibujar circulos costados
#circulo izquierda
cv2.circle(black_board,center=(150,512), radius=100, color=(255,0,0), thickness=-1)

#circulo derecha
cv2.circle(black_board,center=(874,512), radius=100, color=(255,0,0), thickness=-1)

plt.imshow(black_board)
```

```python
#rectangulo superior
cv2.rectangle(black_board,pt1=(300,200),pt2=(724,250),color=(255,0,0),thickness=-1)

#rectangulo inferior
cv2.rectangle(black_board,pt1=(300,824),pt2=(724,774),color=(255,0,0),thickness=-1)

plt.imshow(black_board)
```
