```python
#VARIABLES X y Y

x = dataset.iloc[:,3:13].values #CrediScore, Georphy, Gender,...,Estimated Salary.
y = dataset.iloc[:,13].values #Exited
```

```python
#ETIQUETADO

#Codificacion y categorizacion de las variables independientes.
#LabelEncoder, etiquetado de los datos.
#OneHotEncoder, para la dumificacion de la categorizacion.
from sklearn.preprocessing import LabelEncoder
```

```python
#Etiquetado de las columnas Geography y Gender
labelenconder_x_1 = LabelEncoder()
x[:,1] = labelenconder_x_1.fit_transform(x[:,1]) #Geography
labelencoder_x_2 = LabelEncoder()
x[:,2] = labelencoder_x_2.fit_transform(x[:,2]) #Gender
```

```python
pd.DataFrame(x).head()
```

```python
#DUMIFICACION

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
#Columna 1 'Geography'
ct=ColumnTransformer(transformers=[('oh', OneHotEncoder(),[1])],remainder='passthrough')

#Ajutar y guardar la transformacion en la misma x
x=ct.fit_transform(x)
```

```python
#Se aumentaron mas columnas por la categoriacion de la columna 1 'Geography'
pd.DataFrame(x).head()
```

```python
#OneHotEncoder, siempre eliminar una columna para evitar la multicolineidalidad.
x = x[:,1:]
```

```python
pd.DataFrame(x).head()
```

```python
#DIVISION CUNJUNTO DE DATOS

#Dividir el dataset en conjunto de entrenamiento y pruebas.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
#ESCALADO

#Escalado de datos.
#Para que no exista el sesgo a una determinada columna durante el entrenamiento.
#En redes neuronales artificiales simpre se tiene que estandarizar. 
#En la estandarizacion los datos de ajusta a la media=0 y desviacion estandar=1
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train) #fit_transform
x_test = sc_x.transform(x_test)#solo transform
```

### Construccion de la Red Neuronal Artificial

```python
#keras,es una biblioteca de redes neuronales.
#keras automaticamente usa tensorflow.
#keras has sido concebido para actuar como una interfaz en lugar de ser un framework de machine learning.
#keras ofrece un conjunto de abstracciones mas intiuitivas haciendo mas sencillo el desarrollo de modelos de deep learning.
#Despues de aprender RNA y RNC aprender TensorFlow. Primero v1, luego v2 de TensorFlow.

import keras
from keras.models import Sequential #Inicializacion de parametros de la RNA.
from keras.layers import Dense #Para declarar y crear las capas ocultas de RNA.
```

```python
#Ejercicio de clasificacion, quiero saber si un cliente se va o no del banco.
clasificacion = Sequential()
```

```python
#CAPA DE ENTRADA(INPUT_DIM) y PRIMERA CAPA OCULTA

#Los cientificos de datos desiden el numero de capas. Basado en pruebas y error. No existe una receta.
#Dense es la conexion entre capas, zona de la sinapsis.

#units, es el numero de nodos de la capa oculta.
    #Cantidad de columnas de la capa de entrada + Capa de Salida / media aritmetica
    #Se obtiene de la cantidad de columnas de la matriz de caracteristicas. 11 columnas + 1 salida = 12/2 = 6
#kernel_initializer, inicializacion de los pesos.
    #La distribucion inicializar uniformente, numeros pequenos cercanos a cero pero nunca cero.
#activation, funcion de activacion.
    #En la capa oculta casi simpre se usa la funcion de activacion RELU.
#input_dim, nro. de nodos de entrada de inicializacion de la RNA, 11 columnas.

clasificacion.add(Dense(units = 6, kernel_initializer = "uniform", activation = "relu", input_dim = 11 ))
```

```python
#SEGUNDA CAPA OCULTA

#units,Para que trabaje con la misma cantidad de neuronas que la misma capa. equivilibrada
#kernel_initializer=uniforme
#activation=RELU.

clasificacion.add(Dense(units = 6, kernel_initializer= "uniform", activation = "relu"))
```

```python
#CAPA DE SALIDA

#units, solo un nodo. 
#kernel_initializer, uniforme, el mas recomendable.
#activation, sigmoid. toma valores entre 0 y 1. Resultado de probabilidad.

clasificacion.add(Dense(units = 1, kernel_initializer = "uniform", activation = "sigmoid" ))
```

```python
#COMPILACION DE LA RNA

#optimezer, se encarga de optimizar y encontrar el conjunto optimo de pesos de la RNA.
    #Encuentra los mejores valores de pesos.
    #adam, agarra lo mejor del algoritmo gradiente descendete por bloques y estocastico.
#loss, parametro de control de perdida entre Y^ y Y. Obtiene la funcion de coste.
    #Como el resultado es 1 o 0 (categorias), usar binary_crossentropy.
#metrics, presicion del entrenamiento y al testo que se va a realizar a la RNA.

clasificacion.compile(optimizer = "adam", loss="binary_crossentropy", metrics = ["accuracy"])
```

```python
#ENTRENAMIENTO

#batch_size=Tamano del bloque (gradiente descendente por bloque). Que trabaje de 10 en 10 registros del dataset.
#epochs=Cantidad de epocas o iteraciones. Mientras mas iteraciones mucho mejor.
    #100k epocas es recomendable.
    #En las primeras 10k epocas no aprendia nada. Posterior, recien va
clasificacion.fit(x_train, y_train, batch_size = 10, epochs=300)

'''
loss, resultado funcion de coste. Con el transcurso de las iteraciones tiene que ir disminuyendo.
accuracy, presicion del modelo, Tiene que ir creciendo.

Epoch 296/300
800/800 [==============================] - 1s 823us/step - loss: 0.3322 - accuracy: 0.8618
'''
```

```python
#MATRIX DE CONFUSION

from sklearn.metrics import confusion_matrix
```

```python
y_pred = clasificacion.predict(x_test)
y_pred
```

```python
#Prediciones mayor al 50%.
y_pred = (y_pred > 0.5)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```
