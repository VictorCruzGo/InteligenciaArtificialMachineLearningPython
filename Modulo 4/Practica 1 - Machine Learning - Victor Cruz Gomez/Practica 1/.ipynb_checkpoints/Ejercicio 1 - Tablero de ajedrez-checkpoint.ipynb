```python
#Importar modulos.
import numpy as np
import matplotlib.pyplot as plt
import cv2 #OpenCV
%matplotlib inline
```

```python
#Crear una imagen con fondo blanco de tamano 1024x1024
white_board = np.zeros(shape=(1024,1024,3),dtype=np.int16)
white_board[:,:,:]=(255,255,255)
```

```python
#Mostrar forma de la imagen
white_board.shape
```

```python
#Mostrar la imagen
plt.imshow(white_board)
```

```python
#Dibujar lineas en el tablero blanco
#lineas verticales
cv2.line(white_board,pt1=(350,800),pt2=(350,0),color=(0,0,0),thickness=40)
cv2.line(white_board,pt1=(650,800),pt2=(650,0),color=(0,0,0),thickness=40)
#lineas horizontales
cv2.line(white_board,pt1=(100,250),pt2=(900,250),color=(0,0,0),thickness=40)
cv2.line(white_board,pt1=(100,550),pt2=(900,550),color=(0,0,0),thickness=40)
plt.imshow(white_board)
```

```python
#Dibujar 'O' en la posiciones indicadas
#cv2.circle(img=white_board, center=(150,150), radius=50, color=(0,0,0), thickness=20)
cv2.putText(white_board,'O',(150,150), 3,5,(0,0,0), 20)
cv2.putText(white_board,'O',(750,150), 3,5,(0,0,0), 20)
cv2.putText(white_board,'O',(150,450), 3,5,(0,0,0), 20)
cv2.putText(white_board,'O',(450,450), 3,5,(0,0,0), 20)

plt.imshow(white_board)
```

```python
#Dibujar 'X' en la posiciones indicadas
cv2.putText(white_board,'X', (450,150), 3,5,(0,0,0), 20)
cv2.putText(white_board,'X', (150,750), 3,5,(0,0,0), 20)
cv2.putText(white_board,'X', (450,750), 3,5,(0,0,0), 20)
cv2.putText(white_board,'X', (750,750), 3,5,(0,0,0), 20)
cv2.putText(white_board,'X', (750,450), 3,5,(0,0,0), 20)
cv2.putText(white_board,'!GANASTE!', (350,950), 3,2,(0,0,0), 5)
plt.imshow(white_board)
```
