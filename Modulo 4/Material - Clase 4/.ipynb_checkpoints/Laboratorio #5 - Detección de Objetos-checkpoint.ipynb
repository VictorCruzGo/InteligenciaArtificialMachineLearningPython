```python
full = cv2.imread(path+"sammy.jpg")
full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)
face = cv2.imread(path+"sammy_face.jpg")
face = cv2.cvtColor(face,cv2.COLOR_BGR2RGB)

#Variable de metodos de comparacion de imagenes.
#Detectan un punto de calor.
#los metodos CCOEFF, CCORR, encuentran puntos de calor verde
#El metodo SQDIFF, encuentra puntos de calor oscuros.

#1.cv2.TM_CCOEFF,
    #Ha detectado la cabeza del cachorro.
#2.cv2.TM_CCOEFFNORMED,
    #Con verificacion normada. Con punto de desfase entre -1 y 1.
    #Si su COEF ha detectado su normado tambien lo va a detectar.
#3.cv2.TM_CCORR,
    #Se equivoco en la deteccion.
    #El punto de calor esta mas abajo.
    #El metodo ha confundido la boca con el collar del cachorro.
#4.cv2.TM_CCORR_NORMED,
    #No se equivoco.
    #El punto de equilibrio fue mas fuerte.
#5.cv2.TM_SQDIFF,
    #Busca punto calor oscuro.
    #De los 6 metodos es el mas sensible.
#6.cv2.TM_SQDIFF_NORMED,
    #Por el punto de equilibrio -1 y 1 encuentra alguna iluminacion respecto a TM_SQDIFF

methods = ['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR','cv2.TM_CCORR_NORMED','cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED']
```

```python
#Recorrer cada metodo de comparacion y encontrar sus resultados.
for m in methods:
    #Copia de la imagen
    full_copy = full.copy()
    
    #Evaluacion del metodo de coparacion. 
    method = eval(m)
    
    #Buscar las coincidencia de la imagen 'fuente.'
    #Aplicar el metodo de comparacion. Es decir buscar la cabeza del cachorro en la imagen completa.
    #Si encuentra la imagen buscada, resaltar un punto de calor en la imagen.
    #El resultado es una matriz, donde cada posicion contiene un numero que indica que tan bueno o malo es el match alli.
    #matchTemplate es muy sensible a la escala.
    #cv2.matchTemplate(imagenFuente o fuente,imagenBuscar o plantilla,algoritmoComparacionImagen).
    res = cv2.matchTemplate(full_copy,face,method)
    
    #Valores minimos-maximos y posiciones de los puntos en la matriz
    #Del resultado solo interesan el minimo y maximo en la matrix, la posicion (x,y) del minimo y maximo
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
    print('min_val',min_val, 'max_val',max_val, 'min_loc',min_loc, 'max_loc',max_loc)
    
    #cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED son metodos sensibles de comparacion de imagenes. Pueden tener exitos al localizar.
    #Necesitamos dos puntos. Encontrar el punto superior izquierdo e inferior derecho
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:        
        top_left = min_loc
    else:
        top_left = max_loc
    
    bottom_right = (top_left[0] + width, top_left[1] + height)
    
    #Crear el rectango para marcar la imagen buscada.
    print('top_left',top_left,'bottom_right',bottom_right)
    cv2.rectangle(full_copy, top_left, bottom_right, 255, 10)
    
    #Numero de celda
    plt.subplot(121)
    #Mostrar la imagen
    plt.imshow(res)
    plt.title("Resultado de Verificacion de la Plantilla")
    
    #Numero de celda
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title("Punto Detectado")
    #Colocar en el subtitulo en nombre del metodo de comparacion.
    plt.suptitle(m)
    
    plt.show()
    print("\n")
    print("\n")
    
    
    #Para la deteccion y seguimiento, existen algoritmos preentrenados.
```
