```python
#Fusionar imagenes (addWeighted=Agregar ponderado)

#Para el fusionado las imagenes tienen que tener el mismo tamano.
#alpha,transparencia de la imagen 1
#beta, transparencia de la imagen 2
#gamma=0, nivel de funsionado normal

#blended = cv2.addWeighted(src1=img1, alpha=0.5, src2=img2, beta=0.5, gamma=0)
blended=cv2.addWeighted(src1=img1,alpha=0.5,src2=img2,beta=0.5,gamma=0)
```

```python
#plt.imshow(blended)
plt.imshow(blended)
```

```python
#Fusionado variando la transparencia de las imagenes.

#blended = cv2.addWeighted(src1=img1, alpha=0.1, src2=img2, beta=0.8, gamma=0)
#plt.imshow(blended)

blended=cv2.addWeighted(img1,0.1,img2,0.8,0)
plt.imshow(blended)
```
