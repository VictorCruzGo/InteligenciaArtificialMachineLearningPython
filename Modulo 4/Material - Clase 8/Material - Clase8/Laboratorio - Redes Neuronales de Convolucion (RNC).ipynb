```python
#Instalar Keras. Modulo para implementar redes neuronales.
from keras.models import Sequential #Libreria que permite inicializar la RNC con los pesos cercanos a ceros.
from keras.layers import Conv2D #Libreria para crear la capa de convolucion, filtro, mapa de caracteristicas.
from keras.layers import MaxPooling2D #Libreria para crear las matrices pooleadas.
from keras.layers import Flatten #Libreria para el aplanamiento de datos.
from keras.layers import Dense #Libreria para crear las capas de la RNA
```

```python
#Inicializar la RNC

#Problema de clasificacion. en que porcentaje es perro o gato.
#Imagenes de perros y gatos. 
classifier = Sequential()
```

```python
#1. Capa de convolucion

#filters=32, filtros, refiere a la cantidad de DETECTORES DE RASGOS.
    #Por cada imagen de entrada de la red neuronal
    #Para la generacion de mapas de caracteristicas.
    #32 es un numero generico para creacion de mapas de caracteristicas. 32 matrices de caractarsiticas.
    #Menos de 32 no es eficiente.
    #filters=64,128,256,512,... valores al cuadrado.
    #Cuanto mas mapas de caracteristicas, mas tarda en el entrenamiento, aprende mejor.
#kernel_size = (3,3), tamano del DECTOR DE RASGO.
    #3x3 el tamno de la matriz del detector de rasgos.
#input_shape=(64,64,3), tamano y cantidad de canales de color de las IMAGENES DE ENTRADA. Datos genericos.
    #Lo mas logico es trabajar con 3 canales de color. Son imagenes.
    #En el input shape todas las imagenes van a tener el tamano de 64x64.
    #El tamano de cada imagen puede ser mas grande, no hay ningun problema. Tarda mas horas y mayor proceso.
#activation="relu", para evitar la linealidad, elimnar lineas de cada imagenes. Aplicar RELU al MAPA DE CARACTERISTICAS.

classifier.add(Conv2D(filters=32,kernel_size = (3,3),input_shape=(64,64,3),activation="relu"))
```

```python
#2. Capa de Max Pooling

#Tomar de la imagen original informacion y obtener el valor maximo de cada ventana de 2x2.
#pool_size=(2,2), tamano de la matrix de la etapa de max pooling.
classifier.add(MaxPooling2D(pool_size=(2,2)))
```

```python
#3. Flattening

#Aplanamiento de los datos
classifier.add(Flatten())
```

```python
#4. Full Conection

#Capa de entrada y Capa oculta al mismo tiempo.
#units=128, cantidad de neuronas de entrada.
    #Para tener una conexion del tipo fullconection trabajar con 128 neuronas.
#activation="relu". En la capa oculta siempre trabajar con RELU
classifier.add(Dense(units=128,activation="relu"))

#Capa Salida
#units=1, solo una neurona de salida.
#activation="sigmoid", para entregar resultado de probabilidad
classifier.add(Dense(units=1,activation="sigmoid"))
```

```python
#Compilacion

#Compilacion de la red neuronal de convulcional. Comprobar que las capas y parametros esten correctamente definidos.
#optimizer="adam", fusion del proceso estocastico y bloques. Descenso de gradiente
#loss="binary_crossentropy", al ser un problema de clasificicaion (Perros o Gatos) binario.
#metrics=["accuracy"], nivel de presicion del entrenamiento.
classifier.compile(optimizer="adam",loss="binary_crossentropy", metrics=["accuracy"])
```

```python
#Ajustar la RNC

#Ajustar la RNC a las imagenes de entrenamiento que tenemos que introducir a la red neuronal.
from keras.preprocessing.image import ImageDataGenerator #Generado de datos de imagenes
```

```python
#Configuracion para la generacion de datos de entrenamiento

#Procesamiento de imagenes.
#rescale=1./255, transforma los pixeles con valores entre 0 y 1 flotante.
    #Similar a la normalizacion/estandirizacion.
    #Cada pixel va de 0 a 255.     
#shear_range=0.2, rango de presicion de rescalaje.
    #va de 0.2 a 0.7
#zoom_range=0.2, rango de ampliaje de caja imagenes.
#horizontal_flip=True, volte de la imagen en el eje horizontal.
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)
```

```python
#Configuracion para la generacion de datos de testeo.

#rescale=1./255, solo interesa el rescalado de la imagen
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
#Cargado de imagenes de entrenamiento utilizando la configuracion.

#"dataset/training_set", dentro tiene que existir dos carpetas: gatos y perros
    #Automaticamente se va a reconocer las carpetas.
#target_size=(64,64), tamano de la imagen que se espera cargar.
#batch_size=32, tamano del lote. Tamano de las MAPAS DE CARACTERISTICAS que se espera generar por cada imagen.
#class_mode="binary", modo de catalogacion que se va a realizar, binaria.

training_dataset = train_datagen.flow_from_directory("dataset/training_set",
                                                    target_size=(64,64),
                                                    batch_size=32,
                                                    class_mode="binary")
```

```python
#Cargado de imagenes de entrenamiento utilizando la configuracion.

#"dataset/test_set", directorio donde se encuentran las imagenes de testeo
#target_size=(64,64),
#batch_size=32, mapa de caracteristicas
#class_mode="binary"
#Con los directorios se puede trabajar con rutas relativas y absolutas.
testing_dataset = test_datagen.flow_from_directory("dataset/test_set",
                                                  target_size=(64,64),
                                                  batch_size=32,
                                                  class_mode="binary")

#Carga cada imagen y de da cada imagen realiza el redimensionado
```

```python
#Entrenamiento

#training_dataset, dataset entrenamiento
#steps_per_epoch=8000, cantidad de iteraciones de cada epoca.
    #Cuantas muestras(imagenes) debe tomar dentro de un clico completo de entrenamamiento. 
    #Debe tomar las 8000 imagenes.
    #En cada epoca toma 8000 imagenes.
#epochs=25, numero de epocas.
    #25 es un valor generico de epocas. 
#validation_data=testing_dataset, trabaja con el dataset de testeo
#validation_steps=2000, numero de imagenes en el testeo.
classifier.fit_generator(training_dataset,
                        steps_per_epoch=8000,
                        epochs=25,
                        validation_data=testing_dataset,
                        validation_steps=2000)

#Cuantas epocas? R. Cuantas mas epocas mejor aprende la red nueronal.
#En una sola epoca no se ve toda la foto. En la segunda epoca detecta otros detalles, asi sucesivamente.
#En cada epoca (en cada pasada) el 'accuracy' mejora.
#Muchos trabajan con 30 epocas.


#Epoch 1/25, primera epoca
#ETA: 44:42, aproximadamente cada epoca tardar 44.42 minutos. 
#loss: 0.8524, nivel de perdida, tiene que ser un valor pequeno que se aproxime en cero.
#accuracy: 0.5089, nivel de aprendizaje, tiene que ser arriba del 85%.
```
