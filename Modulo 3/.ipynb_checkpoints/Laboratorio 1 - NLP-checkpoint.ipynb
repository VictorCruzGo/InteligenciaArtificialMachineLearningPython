```python
#frecuencias de palabra
from bs4 import BeatifulSoup
import urllib.request
response=urllib.request.urlopen('http://php.net')
html=response.read() #Leer la respuesta

#html5lib, para que entiendad que es html
soup=BeatifulSoup(html,"html5lib")
#strip=True, es la division que se aplica al texto. Elimina el codigo html
text=soup.get_text(strip=True)

#Dividir todo el texto en palabras
#Cada token es una palabra. 
tokens=[t for t in text.split()]

#frecuencias de palabra
freq=nltk.FreqDist(tokens)

#Palabra repetidas en el texto
for key,val in freq.items():
    print(str(key)+':'+str(val))
```

```python
#cumulative=False, para que no se acumule toda la informacion
freq.plot(20, cumulative=False)
```

```python
#Punto de paradas.
#Siempre eliminar los puntos de paradas.
from nltk.corpus import stopwords
```

```python
#Todo el texto del sitio php es en Ingles
stopwords.words('english')
```

```python
#Almacenando todos los tokens
clean_tokens = tokens[:]
sr=stopwords.words('english')
for token in tokens:
    if token in stopwords.words('english'):
        clean_tokens.remove(token) #limpiando.
```

```python
#Verificacion
freq=nltk.FreqDist(clean_tokens)
for key, val in freq.items():
    print(str(key)+':'+str(val))
```

```python
freq.plot(20, cumulative=False)
```
