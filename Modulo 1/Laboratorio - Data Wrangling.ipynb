```python
#Con mas datos se ve mucho mejor la campana de gauss
#Metodo de montecarlo?  Se utiliza para poder aproximar expresiones matematicas complejas.
#Se usa para trabajar con problemas de aproximacion.

%matplotlib inline
data=np.random.randn(10000)
plt.hist(data)
```

```python
#Generar atraves del metodo de montecarlo el numero PI.
#Ej. cual es la probabilidad de que el dardo caiga dentro o fuera del circulo.
pi_avg=0
n=1000 #limite igual a mil vueltas
pi_value_list=[]

for i in range(100): #Para poder obtener casos favorables de resolucion.
    value=0
    x=np.random.uniform(0,1,n).tolist()
    y=np.random.uniform(0,1,n).tolist()
    for j in range(n):
        z=np.sqrt(x[j]*x[j]+y[j]*y[j])
        if z<=1:
            value+=1
    float_value=float(value)
    pi_value=float_value*4/n
    pi_value_list.append(pi_value)
    pi_avg+=pi_value
    
pi=pi_avg/100

print(pi)

plt.plot(pi_value_list)
    
```
