```python
#Se puden generar graficos.(torta, barras, etc)
#El objetivo de matplotlib es solo mostrar graficos.
```

```python
#Importar la libreria pyplot de matplotlib
#import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
#Crear la lista 'a' y 'b'
#a=[1,2,3,4]
#b=[11,22,33,44]
a=[1,2,3,4]
b=[11,22,33,44]
```

```python
#plot, permite generar una linea. a y b se correlacionan.
#plt.plot(a,b,color='blue',linewidth=3,label='Linea')
#plt.legend()
#plt.show()

plt.plot(a,b, color='red', linewidth=5, label='Linea')
plt.legend()
plt.show()
```
